```python
import numpy as np
import pandas as pd


df = pd.read_csv('data.csv', encoding='latin1', low_memory=False)
x = df['no2']

x = x.dropna()
```

```python
a_r = 0.05 * (102303871 % 7)
b_r = 0.3 * (102303871 % 5 + 1)

z = x + a_r * np.sin(b_r * x)
```

```python
u = np.mean(z)
l = 1/ (2 * np.var(z))
c = np.sqrt(l / np.pi)

print(f' mu = {u} \n lambda = {l} \n c = {c} ')
```

```python
import matplotlib.pyplot as plt

z_vals= np.linspace(0, 900, 500)
pdf= c*np.exp(-l*(z_vals-u)**2)

plt.hist(z, 50, density=True, alpha=0.9, label='histogram of z')
plt.plot(z_vals, pdf, color = 'red', label='est. PDF')
plt.xlabel('z')
plt.ylabel('prob density')
plt.title('NO2 transformed distribution with estimated PDF')
plt.legend()
plt.show()
```
